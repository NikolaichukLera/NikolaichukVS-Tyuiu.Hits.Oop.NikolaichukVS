### Вариант задания 2


<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Student в C#, который будет представлять информацию о студентах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма.

Требования к базовому классу Student:

• Атрибуты: Имя (Name), Возраст (Age), Курс (Course).

• Методы:

o GetInfo(): метод для получения информации о студенте в виде строки.

o Study(): метод для вывода сообщения о том, что студент учится.

o TakeExam(): метод для вывода сообщения о сдаче экзамена.

Требования к производным классам:

Студент бакалавриата (BachelorStudent): Должен содержать дополнительные атрибуты, такие как Специальность (Specialty). Метод Study() должен быть переопределен для добавления информации о специальности при изучении предмета.

Студент магистратуры (MasterStudent): Должен содержать дополнительные атрибуты, такие как Научный руководитель (ScientificAdvisor). Метод TakeExam() должен быть переопределен для добавления информации о научном руководителе при сдаче экзамена.

Аспирант (PhDStudent) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Тема диссертации (DissertationTopic). Метод Study() должен быть переопределен для добавления информации о теме диссертации при изучении предмета.



#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

----

```
using System;

public interface IProject
{
    string Project { get; set; }
    void WorkOnProject();
}

public interface IMemoir
{
    void PublishMemoir();
}

public class Student
{
    public string Name { get; set; }
    public int Age { get; set; }
    public int Course { get; set; }
    public double AvgPoints { get; set; }
    public string University { get; set; }


    public Student(string name, int age, int course, double avgPoints, string unik)
    {
        Name = name;
        Age = age;
        Course = course;
        AvgPoints = avgPoints;
        University = unik;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, возраст: {Age}, курс: {Course}, средний балл: {AvgPoints}, университет: {University} ";
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится.");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен");
    }

    public void FamiliarStudents(Student familiarStudent)
    {
        Console.WriteLine($"{Name} знаком с {familiarStudent.Name}.");
    }
}

class BachelorStudent : Student, IProject
{
    public string Speciality { get; set; }
    public string Project { get; set; }

    public BachelorStudent(string name, int age, int course, string speciality, double avgPoints, string unik, string project) : base(name, age, course, avgPoints, unik)
        {
            Speciality = speciality;
            Project = project;
        }
        
    public override void Study()
    {
        Console.WriteLine($"{Name} учится по специальности {Speciality}.");
    }

    public void WorkOnProject()
    {
        Console.WriteLine($"У {Name} проект: {Project}");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $" специальность: {Speciality}, проект: {Project}";
    }
}

class MasterStudent : Student, IProject, IMemoir
{
    public string ScientificAdvisor { get; set; }
    public string Project { get; set; }

    public MasterStudent(string name, int age, int course, string advisor, double avgPoints, string unik, string project) : base(name, age, course, avgPoints, unik)
        {
            ScientificAdvisor = advisor;
            Project = project;
        }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен у научного руководителя {ScientificAdvisor}, ");
    }

    public void WorkOnProject()
    {
        Console.WriteLine($"У {Name} проект {Project}.");
    }

    public void PublishMemoir()
    {
        Console.WriteLine($"{Name} публикует статью.");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"научный руководитель {ScientificAdvisor}, проект {Project}";  
    }

}

class PhDStudent : Student, IMemoir
{
    public string DissertationTopic { get; set; }
    public int PubMemoir { get; set; }

    public PhDStudent(string name, int age, int course, string topic, double avgPoints, string unik, int memoir) : base(name, age, course, avgPoints, unik)
        {
            DissertationTopic = topic;
            PubMemoir = memoir;
        }

    public override void Study()
    {
        Console.WriteLine($"{Name} работает над диссертацией: {DissertationTopic},");
    }

    public void PublishMemoir()
    {
        Console.WriteLine($"{Name} опубликовал {PubMemoir} статьи.");
    }

    public override string GetInfo()
    {
        return base.GetInfo() +$" тема диссертации: {DissertationTopic}, кол-во публикаций: {PubMemoir}";
    }
}

BachelorStudent bachelor = new BachelorStudent("Александр Груздин", 18, 1, "Бизнес информатика", 3.55, "ВШЦТ", "Как завлечь клиентов");
MasterStudent master = new MasterStudent("Константин Онисимов", 22, 2, "Олег Дмитриевич", 4.22, "ИГИН", "Потребительское мышление в бизнесе");
PhDStudent phds = new PhDStudent("Анфиса Чехова", 20, 3, "Информатика и вычислительная техника", 4.9, "АРХиД", 2);

bachelor.Study();
Console.WriteLine(bachelor.GetInfo());
bachelor.TakeExam();
bachelor.WorkOnProject();

Console.WriteLine("-------------------------------------");

master.Study();
Console.WriteLine(master.GetInfo());
master.TakeExam();
master.WorkOnProject();
master.PublishMemoir();

Console.WriteLine("-------------------------------------");

phds.Study();
Console.WriteLine(phds.GetInfo());
phds.TakeExam();
phds.PublishMemoir();

Console.WriteLine("-------------------------------------");

bachelor.FamiliarStudents(master);
master.FamiliarStudents(phds);
```
